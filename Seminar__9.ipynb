## Задача №67.
1. Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.

high - длинный(от 42)

middle - средний(от 35 до 42)

low - маленький(до 35)


```python
df.sample()    # Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.
```

```python
# 2. high - длинный(от 42)  'height_group' создали новую колонку, куда будем записывать
df.loc[df['bill_length_mm'] >= 42, 'heigh_group']= 'heigh'
# 3. middle - средний(от 35 до 42)
df.loc[(df['bill_length_mm'] >= 35) & (df['bill_length_mm'] < 42), 'heigh_group'] = 'middle'
# 4. low - маленький(до 35)
df.loc[df['bill_length_mm'] < 35, 'heigh_group']= 'low'
df.sample(40)  # (размер строк)
```
